函数（function）

```python
def add_list(num_list):
    sum = 0
    for num in num_list:
        sum += num
    return sum

result = add_list([1, 2])
print(result)
```

```python
for i in range(1, 101):
    num_list = list(range(i + 1))
    sum = add_list(num_list)
    print(sum, num_list)
```

```python
def add_1_to_n(n):
    count = 1
    sum = 0
    while count <= n:
        sum += count
        count += 1
    return sum


sum = add_1_to_n(500)
print(sum)
```

```python
def n_m_3_5(n):
    my_list = []
    count = 1
    while count <= n:
        if count % 3 == 0 and count % 5 == 0:
            print(count)
            my_list.append(count)
        count += 1
    
    return my_list

my_list = n_m_3_5(30)
print(f"my_list:{my_list}")
```

```python
def check_input(input):
    if input == 'q':
        print("Quit")
        return 1
    if input.isalpha():
        print("Please input a number!!!")
        return 2
    number = int(input)
    if number > 0:
        print("It is a positive number.")
    elif number < 0:
        print("It is a negetive number.")
    else:
        print("It is zero.")

    return 0
        

while True:
    number = input("Please input a number:")
    
    return_code = check_input(number)
    if return_code == 1:
        break

    
print("code after loop.")
```

```python
def draw_diamonds(rows):
# rows = 7
    for i in range(rows):
        space = abs(rows // 2 - i)
        star = rows - 2 * space
        print(f"{' ' * space}{'*' * star}")

draw_diamonds(15)
```
