读操作

```python
with open('file.txt', 'rt') as f:
    content = f.readlines()
print(type(content), len(content), content)
# f.close()
```

写操作

```python
with open("temp.txt", 'w') as f:
    original_content = ['Open file and return a corresponding file object.', 'If the file cannot be opened, an OSError is raised.', 'See Reading and Writing Files for more examples of how to use this function.']
    modified_content = [f"{content}\n" for content in original_content]
    f.writelines(modified_content)
```
